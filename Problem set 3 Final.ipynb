# Question 6

Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
marriage = pd.read_csv('D:\\Assignments\\data_programming\\us-marriages-divorces-1867-2014.csv')
```

```python
marriage.head(10)
```

```python
# get columns to plot
columns = marriage.columns.drop(['Year'])
# create x data
x_data = range(0, marriage.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, marriage[column])
# set title and legend
ax.set_title('Divorce')
ax.legend(columns)
```

# Question 7

Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
plt.figure(figsize = (22, 8))
plt.bar(x =marriage['Year'],
       height = marriage['Marriages'],
       color = 'Navy')
plt.xticks(rotation = 90, fontsize = 15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Divorces', fontsize= 11);
plt.title('Marriages and Divorces per capita in the U.S. between 1900, 1950, and 2000', fontsize=16, color = 'Black',fontweight='bold')
```
